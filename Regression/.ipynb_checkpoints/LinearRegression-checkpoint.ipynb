```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_excel("houseData.xlsx")
df
```

```python
%matplotlib inline
plt.xlabel("area (Sq Ft)")
plt.ylabel("Price (USD)")
plt.scatter(df.area, df.price, color="red", marker="+")
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[["area"]], df.price)
```

```python
reg.predict([[3300], [4000], [5000]])
```

```python
%matplotlib inline
plt.xlabel("area (Sq Ft)")
plt.ylabel("Price (USD)")
plt.scatter(df.area, df.price, color="red", marker="+")
plt.plot(df.area, reg.predict(df[["area"]]), color="blue")
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
areas = pd.read_excel("areas.xlsx")
areas
```

```python
predicted = reg.predict(areas)
```

```python
areas["predicted_values"] = predicted
```

```python
areas.to_csv("predictedValues.csv", index=False)
```
